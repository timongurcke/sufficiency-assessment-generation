```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

pd.set_option('display.max_colwidth', None)

df_aae_full = pd.read_csv('../../../ceph_data/input/UKP-InsufficientArguments_v1.0/data-tokenized.tsv', sep='\t', index_col=False, encoding='latin-1')
df_aae = pd.read_json('../../../ceph_data/intermediate/corpus-ukp-argument-annotated-essays-v2/aae3.json')
df_aae_invalid = pd.read_json('../../../ceph_data/intermediate/corpus-ukp-argument-annotated-essays-v2/aae3_invalid.json')
df_aae_fixed = pd.read_json('../../../ceph_data/intermediate/corpus-ukp-argument-annotated-essays-v2/aae3_fixed.json')
df_split = pd.read_csv('../../../ceph_data/input/UKP-InsufficientArguments_v1.0/data-splitting.tsv', sep='\t', names=['index']+[str(i) for i in range(100)], index_col=False)

df_aae_full['index'] = df_aae_full.apply(lambda x: 'essay{}_{}'.format(str(x['ESSAY']).zfill(3), x['ARGUMENT']), axis=1)
df_aae_full['local_sufficency'] = df_aae_full['ANNOTATION'].apply(lambda x: 0 if x == 'insufficient' else 1)
```

```python
df_aae_full = df_aae_full[['index', 'local_sufficency', 'TEXT']]
```

```python
df_aae['au_sep'] = df_aae[['au','conclusion']].apply(lambda x: x[0].replace(x[1], '</s> '+x[1]+ ' </s>'), axis=1)
```

```python
df_aae_invalid = df_aae_invalid[df_aae_invalid['index'].str[-2]=='_']
```

```python
sep_dict = {}
for i, row in df_aae_invalid.iterrows():
    if row['index'] not in sep_dict:
        sep_dict[row['index']] = row['au']
    sep_dict[row['index']] = sep_dict[row['index']].replace(row['conclusion'], '</s> '+row['conclusion']+ ' </s>')
```

```python
df_aae_invalid['au_sep'] = df_aae_invalid['index'].apply(lambda x: sep_dict[x])
```

```python
len(df_aae_invalid[['index','au_sep']].drop_duplicates())
```

```python
df_aae_invalid.head()
```

```python
only_claims = list(set(df_aae_full['index']) - set(df_aae['index']) - set(df_aae_invalid['index']))
```

```python
df_aae_full_only_claims = df_aae_full[df_aae_full['index'].isin(only_claims)]
```

```python
df_aae_full_only_claims['au_sep'] = df_aae_full_only_claims['TEXT'].apply(lambda x: '</s> '+x+ ' </s>')
```

```python
df_aae_full_only_claims.head()
```

```python
valid_indexes = df_aae['index'].tolist()
valid_aus = df_aae['au_sep'].tolist()
invalid_indexes = df_aae_invalid[['index','au_sep']].drop_duplicates()['index'].tolist()
invalid_aus = df_aae_invalid[['index','au_sep']].drop_duplicates()['au_sep'].tolist()
only_claim_indexes = df_aae_full_only_claims['index'].tolist()
only_claim_aus = df_aae_full_only_claims['au_sep'].tolist()

indexes = valid_indexes+invalid_indexes+only_claim_indexes
aus = valid_aus+invalid_aus+only_claim_aus

full_sep_dict = dict(zip(indexes,aus))
df_aae_full['TEXT'] = df_aae_full['index'].apply(lambda x: full_sep_dict[x])
```

```python
df_aae_full['placeholder1'] = np.arange(len(df_aae_full))
df_aae_full['placeholder2'] = np.arange(len(df_aae_full))
df_aae_full['placeholder3'] = np.arange(len(df_aae_full))

path = '../../../ceph_data/intermediate/bert-AAE-v2-only-dot-direct-cola-au-full-sep'
if not os.path.exists(path):
    os.mkdir(path)
else:
    shutil.rmtree(path)
    os.mkdir(path)
    
path2 = '../../../ceph_data/output/bert-AAE-v2-only-dot-direct-cola-au-full-sep'
if not os.path.exists(path2):
    os.mkdir(path2)
else:
    shutil.rmtree(path2)
    os.mkdir(path2)

for i in range(100):
    if not os.path.exists(path+'/{}'.format(i)):
        os.mkdir(path+'/{}'.format(i))
    else:
        shutil.rmtree(path+'/{}'.format(i))
        os.mkdir(path+'/{}'.format(i))
        
    if not os.path.exists(path2+'/{}'.format(i)):
        os.mkdir(path2+'/{}'.format(i))
    else:
        shutil.rmtree(path2+'/{}'.format(i))
        os.mkdir(path2+'/{}'.format(i))
    
    split_dict = dict(zip(df_split['index'], df_split[str(i)]))
    df_aae_full['split'] = df_aae_full['index'].apply(lambda x: split_dict[x])
    df_aae_full[df_aae_full['split']=='TRAIN'][['local_sufficency','placeholder1','placeholder2','TEXT','placeholder3']].to_csv('../../../ceph_data/intermediate/bert-AAE-v2-only-dot-direct-cola-au-full-sep/'+str(i)+'/train.tsv',sep='\t', index=False)
    df_aae_full[df_aae_full['split']=='DEV'][['local_sufficency','placeholder1','placeholder2','TEXT','placeholder3']].to_csv('../../../ceph_data/intermediate/bert-AAE-v2-only-dot-direct-cola-au-full-sep/'+str(i)+'/dev.tsv',sep='\t', index=False)
    df_aae_full[df_aae_full['split']=='TEST'][['local_sufficency','placeholder1','placeholder2','TEXT','placeholder3']].to_csv('../../../ceph_data/intermediate/bert-AAE-v2-only-dot-direct-cola-au-full-sep/'+str(i)+'/test.tsv',sep='\t', index=False)
```
