```python
preds_dict = {}
for i in range(5):
    path = '../../../ceph_data/output/bart-AAE-v2-only-dot-direct-cola-au-full-mask-gen-only-suff/{}/aee.preds'.format(i)
    path2 = '../../../ceph_data/intermediate/bart-AAE-v2-only-dot-direct-cola-au-full-mask-gen-only-suff/{}/test_and_insufficient.target'.format(i)
    with open(path, encoding='utf-8') as h:
        preds = h.readlines()
    with open(path2, encoding='utf-8') as f:
        gts = f.readlines()
    for pred, gt in zip(preds,gts):
        preds_dict[gt.replace('\n','')]=pred.replace('\n','')

df_aae_full = pd.read_csv('../../../ceph_data/input/UKP-InsufficientArguments_v1.0/data-tokenized.tsv', sep='\t', index_col=False, encoding='latin-1')
df_aae = pd.read_json('../../../ceph_data/intermediate/corpus-ukp-argument-annotated-essays-v2/aae3.json')
df_aae_invalid = pd.read_json('../../../ceph_data/intermediate/corpus-ukp-argument-annotated-essays-v2/aae3_invalid.json')
df_aae_fixed = pd.read_json('../../../ceph_data/intermediate/corpus-ukp-argument-annotated-essays-v2/aae3_fixed.json')
df_split = pd.read_csv('../../../ceph_data/input/UKP-InsufficientArguments_v1.0/data-splitting.tsv', sep='\t', names=['index']+[str(i) for i in range(100)], index_col=False)
df_aae_full['index'] = df_aae_full.apply(lambda x: 'essay{}_{}'.format(str(x['ESSAY']).zfill(3), x['ARGUMENT']), axis=1)
df_aae_full['local_sufficency'] = df_aae_full['ANNOTATION'].apply(lambda x: 0 if x == 'insufficient' else 1)

df_aae_full_mask = pd.read_csv('df_aae_full_mask.csv')

only_claims = list(set(df_aae_full['index']) - set(df_aae['index']) - set(df_aae_invalid['index']))
df_aae_full_only_claims = df_aae_full[df_aae_full['index'].isin(only_claims)]

conclusions_dict = {}
for i, row in df_aae.iterrows():
    if row['index'] not in conclusions_dict:
        conclusions_dict[row['index']] = []
    conclusions_dict[row['index']].append(row['conclusion'])
for i, row in df_aae_invalid.iterrows():
    if row['index'] not in conclusions_dict:
        conclusions_dict[row['index']] = []
    conclusions_dict[row['index']].append(row['conclusion'])
for i, row in df_aae_full_only_claims.iterrows():
    if row['index'] not in conclusions_dict:
        conclusions_dict[row['index']] = []
    conclusions_dict[row['index']].append('')

replaced_text_dict = {}
for index in df_aae_full_mask['index']:
    rel_df = df_aae_full_mask[df_aae_full_mask['index']==index]
    masked_text = None
    j = 0
    for i, row in rel_df.iterrows():
        if masked_text == None:
            masked_text = row['TEXT_MASK']
        #print(conclusions_dict[row['index']])
        #print(j)
        if conclusions_dict[row['index']][j] in preds_dict:
            masked_text = masked_text.replace('<mask>', '<span style="background-color: #d8d2e7;">' + preds_dict[conclusions_dict[row['index']][j]] + '</span>')
        else:
            masked_text = '<span style="background-color: #d8d2e7;">' + row['TEXT'] + '</span>'
        masked_text = masked_text.replace('<unk>', '<mask>')
        j += 1
    replaced_text_dict[index] = '<p>'+masked_text+'</p> <p style="text-align: right;"><span style="color: #d8d2e7;">▇▇▇</span> = claim&nbsp; &nbsp; &nbsp;<span style="color: #fdf2d0;">▇▇▇</span> = premise</p>'
df_aae_full['REPLACED_TEXT'] = df_aae_full['index'].apply(lambda x: replaced_text_dict[x])    
    
```

```python
study_df = pd.DataFrame(data={'index': sep_df['index'], 'local_sufficency': sep_df.local_sufficency, 'gt': sep_df.TEXT, 'supervised': df_aae_full.REPLACED_TEXT, 'unsupervised': filled_df.REPLACED_TEXT, 'premises': sep_df.premises})
```

```python
def replace_premises(text, premises):
    for premise in premises:
        text = text.replace(premise, '<span style="background-color: #fdf2d0;">'+premise+'</span>')
    return text
```

```python
study_df['gt'] = study_df[['gt','premises']].apply(lambda x: replace_premises(x[0], x[1]), axis=1)
study_df['supervised'] = study_df[['supervised','premises']].apply(lambda x: replace_premises(x[0], x[1]), axis=1)
study_df['unsupervised'] = study_df[['unsupervised','premises']].apply(lambda x: replace_premises(x[0], x[1]), axis=1)
```

```python
study_df = study_df.sample(frac=1, random_state=23).drop_duplicates(subset=['supervised'])
sample_df = study_df.groupby('local_sufficency').apply(lambda x: x.sample(n=50))
sample_df = sample_df.sample(frac=1).reset_index(drop=True)
```

```python
rng = np.random.default_rng()
shuffle = []
for i in range(100):
    arr = [0,1,2]
    rng.shuffle(arr)
    shuffle.append(arr)
shuffle = np.array(shuffle)

shuffled_sample_tgts = np.array([sample_df[['gt','supervised','unsupervised']].values[i][shuffle[i]] for i in range(100)])
sample_df['first'] = shuffled_sample_tgts[:,0]
sample_df['second'] = shuffled_sample_tgts[:,1]
sample_df['third'] = shuffled_sample_tgts[:,2]

sample_df['id1'] = shuffle[:,0]
sample_df['id2'] = shuffle[:,1]
sample_df['id3'] = shuffle[:,2]
```
